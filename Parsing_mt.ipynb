```python
import pandas as pd
data = pd.read_excel("Ozon.xlsx")
```

```python
data = data[~data["Описание"].isna()]
```

```python
data = data[["Характеристики","Описание"]]
data.index = range(data.shape[0])
```

```python
a = ord('а')
alphabet =''.join([chr(i) for i in range(a,a+32)])
def parse_ch(inp: str) -> str:
  flag = False
  text = ""
  for i in inp:
    if i == ":":
      flag = True
    if i == "|" and flag == True:
      flag = False
    if flag == True and i != ":" and i.lower() in alphabet or i == " ":
      text += i
    if flag == True and i != ":" and i.lower() not in alphabet:
      text += " "
  return " ".join([j if len(j) != 1 else "" for j in text.split()])
```

```python
def parse_description(inp):
  flag = True
  text = ""
  for i in inp:
    if flag == True and i != ":" and i.lower() in alphabet + "ё"  or i == " ":
      text += i
    if flag == True and i != ":" and i.lower() not in alphabet:
      text += " "
  return " ".join(text.split())
```
